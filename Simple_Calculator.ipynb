```
# Simple Calculator 
def Addition():
  print("Addition")
  n=float (input("Enter a number: "))
  t=0
  a=0
  while n!=0 :
   a=a+n
   t=t+1
   n=float (input("Enter another number (0 to calculate): "))
  return [a,t]

def Subtraction():
  print("Subtraction")
  n=float (input("Enter a number: "))
  t=0
  a=0
  while n!=0 :
   a=n-a
   t=t+1
   n=float (input("Enter another number (0 to calculate): "))
  return [0-a,t]

def Multiplication():
  print("Multiplication")
  n=float (input("Enter a number: "))
  t=0
  a=1
  while n!=0 :
   a=a*n
   t=t+1
   n=float (input("Enter another number (0 to calculate): "))
  return [a,t]

def Division ():
  print("Division")
  n=float (input("Enter a number: "))
  a=float (input("Enter another number: "))
  ans=n/a
  t=2
  return [ans,t]

def Modulo ():
  print("Remainder")
  n=float (input("Enter a number: "))
  a=float (input("Enter another number: "))
  ans=n%a
  t=2
  return [ans,t]

def Average():
  print("Average")
  an=[]
  an = Addition ()
  sum = an[0] 
  t= an[1] 
  a=sum/t
  return [a,t]

while True :
  list = []
  print("This is a Simple Calculator")
  print("Enter 'a' for Addition")
  print("Enter 's' for Subtraction")
  print("Enter 'm' for Multiplication")
  print("Enter 'd' for Division")
  print("Enter 'r' for Remainder")
  print("Enter 'v' for Average")
  print("Enter 'q' for quit")
  
  c = input(" ")
  
  if c != 'q':
   if c =='a':
    list = Addition()
    print("Ans = ", list[0], " total inputs ",list[1])
   elif c =='s':
    list = Subtraction()
    print("Ans = ", list[0], " total inputs ",list[1])
   elif c =='m':
    list = Multiplication()
    print("Ans = ", list[0], " total inputs ",list[1])
   elif c == 'd':
    list = Division()
    print("Ans = ", list[0], " total inputs ",list[1])
   elif c =='r':
    list = Modulo()
    print("Ans = ", list[0], " total inputs ",list[1])
   elif c =='v':
    list = Average()
    print("Ans = ", list[0], " total inputs ",list[1])
   else :
     print("Invalid Choice!")
  else :
    print("The End!!")
    break



```
